## Question 1

### 1. Simple Linear Regression vs. Multiple Linear Regression
- The difference between Simple Linear Regression and Multiple Linear Regression is that Simple Linear Regression (SLR) models the relationship between a single predictor (independent) variable and a response (dependent) variable. In contrast, Multiple Linear Regression (MLR) incorporates multiple predictors to model the response variable. 
- The benefit the latter provides over the former is that by including multiple predictors, MLR captures more complexity, allowing it to explain variations in the response variable better than SLR, particularly when multiple factors affect the outcome.
- Linear Forms:
    - Simple Linear Regression (SLR):
        - y = β0 + β1x + ϵ
        - y is the response variable, x is a single predictor, β0 is the intercept, β1 is the slope (representing the change in y for a one-unit change in x), and ϵ is the error term. This form assumes that y varies linearly with x alone
    - Multiple Linear Regression (MLR):
        - y = β0 + β1x1 + β2x2 + ... + βnxn + ϵ
        - y depends on multiple predictors x1, x2, ..., xn, each with its own coefficient βi. The presence of multiple predictors allows MLR to model the effect of several variables on y, potentially increasing the explanatory power over SLR by capturing more complex relationships.

### 2. Continuous vs. Indicator Variables in Simple Linear Regression
-  Continuous Variable: A continuous predictor represents a measurable quantity, such as age or height, where values are on a scale. It creates a linear form where the slope reflects the rate of change in the response per unit change in the predictor.
    - Linear Form (Continuous Variable (SLR with Continuous x)):
        - y = β0 + β1x + ϵ
        - x is a continuous predictor, and β1 is the slope, representing the continuous effect on y for each unit change in x. The model’s prediction line has a single slope and intercept.
- Indicator Variable: This binary variable, typically coded as 0 or 1, represents categories (e.g., male/female). The linear form changes the intercept without affecting the slope, indicating the mean difference between the two categories.
    - Linear Form (Indicator Variable (SLR with Indicator z)):
        - y = β0 + β1z + ϵ
        - In this form, z is an indicator variable that takes values 0 or 1, often representing two categories (e.g., 0 for male, 1 for female). The intercept is β0 when z = 0, and β0 + β1 when z = 1. Thus, β1 indicates the difference in mean response between the two categories.

### 3. Effect of Adding an Indicator Variable to Create Multiple Linear Regression
- The change that happens in the behavior of the model (i.e., the expected nature of the data it models) when a single indicator variable is introduced alongside a continuous variable to create a Multiple Linear Regression is that by adding an indicator variable alongside a continuous predictor in MLR, it models an additional categorical effect, creating separate intercepts for each category (without affecting the slope). This enables the model to account for differences between categories while still relating the continuous variable to the response.
- Linear Forms: In SLR, the model has a single line; with MLR, each category has its own line with the same slope but a different intercept.
    - y = β0 + β1x + β2z + ϵ
    - x is a continuous variable, and z is an indicator variable (0 or 1). This form has two intercepts: β0 when z = 0 and β0 + β2 when z = 1. The slope β1 remains the same for both groups, meaning the relationship between x and y is parallel across the categories of z, but each has a unique intercept.

### 4. Effect of Adding an Interaction Term Between Continuous and Indicator Variables in MLR
- The effect of adding an interaction between a continuous and an indicator variable in Multiple Linear Regression model allows the relationship between the continuous predictor and the response to vary across the levels of the indicator variable, meaning each category has both a unique intercept and a unique slope.
- Linear Form: The linear form now includes an interaction term, leading to distinct lines for each category that differ in both slope and intercept, capturing more nuanced relationships.
    - y = β0 + β1x + β2z + β3(x⋅z) + ϵ
    - The interaction term β3(x⋅z) allows the slope of x to vary depending on the value of z. When z = 0, the equation reduces to y = β0 + β1x. When z = 1, it becomes y = (β0 + β2) + (β1 + β3)x, giving a unique intercept and slope for each category of z. Thus, each category has its own line with a distinct slope and intercept, allowing more flexibility in modeling category-specific trends.

### 5. MLR Model Based Only on Indicator Variables from a Non-Binary Categorical Variable:
- The behavior of a Multiple Linear Regression model (i.e., the expected nature of the data it models) based only on indicator variables derived from a non-binary categorical variable is that when an MLR model includes only indicator variables derived from a non-binary categorical variable (e.g., encoding "Color" with categories Red, Green, and Blue), it captures differences in the response variable across these categories.
- Linear Form: This linear form involves several intercepts (one per category, using one as a reference), with each encoding indicating presence/absence of the category. Encodings are binary (0 or 1) and provide flexibility to compare multiple categories, each having a unique effect on the response.
    - y = β0 + β1d1 + β2d2 + ... + βk−1dk−1 + ϵ
    - For a categorical variable with k levels (e.g., colors Red, Green, Blue), we create k − 1 binary indicator variables (e.g., d1, d2) where each variable represents a specific category (1 if present, 0 otherwise). The intercept β0 represents the reference category (e.g., Blue), while β1, β2, ..., βk−1 indicate the mean difference in y from the reference category to each other category. Each binary variable captures the unique effect of its corresponding category, modeling categorical distinctions without assuming any ordering.

I have chosen the following variables:
- Continuous predictor: age (since age is a numeric variable, it will be treated as continuous)
- Binary predictors: gender and residence (these are categorical variables that can be encoded as binary, typically 0 or 1)
- Outcome variable: PSYCH_body_self_image_questionnaire_negative_affect_score (this appears to be a score, which is numeric, so it's treated as the outcome variable)

### Model Formula
The multiple linear regression model you're trying to fit would be:
- PSYCH_body_self_image_questionnaire_negative_affect_score = β0 + β1⋅age + β2⋅gender + β3⋅residence + ϵ
- Where:
    - β0 is the intercept (the baseline value when all predictors are zero)
    - β1 is the coefficient for age, which indicates the change in the outcome variable for each one-unit increase in age
    - β2 is the coefficient for gender, which represents the difference in the outcome variable between the two gender categories (e.g., male vs female).
    - β3 is the coefficient for residence, which represents the difference in the outcome variable between the two residence categories (e.g., urban vs rural).

### Fitting the Model

```python
import pandas as pd
import statsmodels.formula.api as smf

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv"
data = pd.read_csv(url, low_memory=False)

# Check the first few rows to see the relevant columns
print(data[['DEMO_age', 'DEMO_gender', 'GEO_residence_canada', 'PSYCH_body_self_image_questionnaire_negative_affect_score']].head())

# If needed, residence as binary (assuming they are categorical with 'Man'/'Woman' and Yes'/'No')
data['gender_binary'] = data['DEMO_gender'].apply(lambda x: 0 if x == 'Man' else 1)
data['residence_binary'] = data['GEO_residence_canada'].apply(lambda x: 0 if x == 'Yes' else 1)

# Drop rows with missing values
data_clean = data.dropna(subset=['DEMO_age', 'gender_binary', 'residence_binary', 'PSYCH_body_self_image_questionnaire_negative_affect_score'])

# Fit the multiple linear regression model
model = smf.ols('PSYCH_body_self_image_questionnaire_negative_affect_score ~ DEMO_age + gender_binary + residence_binary', data=data).fit()

# View the model summary
print(model.summary())
```

### Model Interpretation:
- Intercept (β0): The predicted score for PSYCH_body_self_image_questionnaire_negative_affect_score when age, gender, and residence are at their baseline values (e.g., age = 0, gender = 0, residence = 0).
- Age Coefficient (β1): The change in the outcome score for each one-unit increase in age. For example, if β1 = 0.5, then for each additional year of age, the PSYCH_body_self_image_questionnaire_negative_affect_score increases by 0.5.
- Gender Coefficient (β2): The difference in the outcome score between the two genders. If β2 = 2, it means females (coded as 1) have a score 2 units higher than males (coded as 0), keeping other variables constant.
- Residence Coefficient (β3): The difference in the outcome score between those living in Canada and those not. If β3 = −1, people living in Canada (coded as 0) have a score 1 unit lower than those living outside of Canada (coded as 1), keeping other variables constant.

### Making Predictions

```python
# Make predictions using the fitted model
predictions = model.predict(data)

# Add predicted outcomes to the dataset
data['predicted_affect_score'] = predictions

# View the first few rows with actual and predicted values
print(data[['PSYCH_body_self_image_questionnaire_negative_affect_score', 'predicted_affect_score']].head())
```

### Visual the Results

### Model Overview:
- The formula defines a model with predictors (Attack, Defense, Speed, Sp. Def, Sp. Atk, Legendary) and their interactions.
- The interaction terms create additional predictor variables that represent combinations of the original predictors (e.g., Attack * Defense, Speed * Legendary).

### Centering and Scaling:
- Centering subtracts the mean, and scaling divides by the standard deviation for each numerical predictor.
- This makes all predictors have a mean of 0 and a variance of 1, which helps normalize the data and reduce potential issues caused by different scales.

### Design Matrix:
- The design matrix is a table where each column corresponds to one of the predictors (including interactions).
- The rows represent observations, and the matrix is used to estimate the outcome variable (HP).

### Multicollinearity:
- Multicollinearity occurs when some predictors are highly correlated with each other, causing redundancy in the information provided by the predictors.
- This issue becomes more prominent with interaction terms, where the relationships between predictors overlap, leading to high correlations.

### Condition Number:
- The "Condition Number" is a measure of the stability of the design matrix. A high condition number (e.g., 2.25 × 10¹⁵) suggests the matrix is ill-conditioned, meaning the model is sensitive to small changes in the data.
- This indicates potential multicollinearity, where the predictors’ effects cannot be reliably distinguished.

### Impact on Model:
- Overfitting: The model fits the training data well but may not generalize to new data because of the unstable coefficients caused by multicollinearity.
- Unstable Coefficients: Small changes in the data can cause large variations in the estimated coefficients, making the model unreliable.
- Poor Generalization: Due to overfitting and instability, the model may perform poorly on out-of-sample data.

### Why Centering and Scaling Don’t Fully Solve the Issue:
- While centering and scaling standardize the predictors, they don't eliminate the underlying correlations between them, especially in complex interaction models.
- Therefore, the model still faces multicollinearity despite these transformations.

```python
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Define the linear model specification
model_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
model_linear_form += ' + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

# Number of iterations
n_iterations = 50  # You can change this depending on how many splits you want to test

# List to collect results
in_sample_r2 = []
out_sample_r2 = []

for i in range(n_iterations):
    # Split the data without setting a fixed seed (so the splits will vary across iterations)
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)

    # Convert categorical variables to have the same categories in both train and test
    for col in ['Type 1', 'Type 2', 'Generation']:
        pokeaman_train[col] = pd.Categorical(pokeaman_train[col])
        pokeaman_test[col] = pd.Categorical(pokeaman_test[col], categories=pokeaman_train[col].cat.categories)

    # Specify and fit the model
    model_spec = smf.ols(formula=model_linear_form, data=pokeaman_train)
    model_fit = model_spec.fit()

    # In-sample R-squared
    in_sample_r2.append(model_fit.rsquared)

    # Out-of-sample R-squared
    yhat = model_fit.predict(pokeaman_test)  # Ensure pokeaman_test is a DataFrame
    y = pokeaman_test.HP
    out_sample_r2.append(np.corrcoef(y, yhat)[0, 1] ** 2)

# Create a DataFrame to hold the results for easy visualization
results_df = pd.DataFrame({
    'In-sample R-squared': in_sample_r2,
    'Out-of-sample R-squared': out_sample_r2
})

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(results_df['In-sample R-squared'], label='In-sample R-squared', color='blue', alpha=0.7)
plt.plot(results_df['Out-of-sample R-squared'], label='Out-of-sample R-squared', color='red', alpha=0.7)
plt.xlabel('Iteration')
plt.ylabel('R-squared')
plt.title('In-sample vs. Out-of-sample R-squared Across Iterations')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

In this demonstration, I’m fitting a linear regression model to a dataset called pokeaman and assessing its performance using two key measures: in-sample R-squared and out-of-sample R-squared.

### 1. Purpose:
- Data splitting: For each iteration, I randomly divide the data into two sets:
    - Training set: This is used to train the model.
    - Test set: This set is used to see how well the model performs on new, unseen data.
- Fitting the model: The model uses features like Attack, Speed, Sp. Def, Sp. Atk, and categorical variables like Generation and Type to predict HP (Hit Points). This is the relationship I’m trying to model.
- Evaluating performance:
    - In-sample R-squared tells me how well the model fits the data it was trained on. A higher value means the model explains a larger portion of the variability in HP within the training set.
    - Out-of-sample R-squared evaluates how well the model generalizes to new data (the test set). It gives me an idea of how accurately the model can predict HP for new, unseen data. A higher out-of-sample R-squared means the model performs better on new data.

### 2. Results:
- In-sample R-squared shows how well the model fits the training data. This value is typically higher because the model is directly trained on this data.
- Out-of-sample R-squared measures the model's ability to make predictions on new data, and it’s usually a bit lower than the in-sample R-squared. This indicates that while the model may fit the training data well, it might not be as good at predicting new data.

### 3. Key Takeaways:
- In-sample vs. Out-of-sample R-squared: If the in-sample R-squared is much higher than the out-of-sample R-squared, it might mean the model is overfitting—it’s too closely tailored to the training data and might not do as well with new data. If both values are close, it suggests that the model generalizes well to unseen data.
- Fluctuations across iterations: Since the data is split randomly each time, the performance can vary across different iterations. That’s why you might see some ups and downs in the plot.
